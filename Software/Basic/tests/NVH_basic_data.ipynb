```python
from scipy.io import loadmat
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np

y = loadmat('Sing_data.mat')
sim = y['simout']
time = sim[0][0][0]
time = time.reshape((1,time.size))
time = time[0]
value = sim[0][0][1][0][0][0]
value = value.reshape((1,value.size))
value = value[0]
print(time)
print(value)


fs = 1/(time[1]-time[0])
f, t, Zxx = signal.stft(value, fs, window='hann', nperseg=512,noverlap =0, )
print(Zxx)
plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=1)
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()


# value = sim[0][0][1]
# arr = value[0][0][0]
# value1 = arr[0][0]
# value2 = arr[1][0]
# value3 = arr[2][0]
# overall = np.concatenate((value1,value2,value3),axis=0)
# print(overall)
```
